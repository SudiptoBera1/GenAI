## Solution - Structured Output

```python
from pydantic import BaseModel,Field
from typing import List

class ResponseStructure(BaseModel):
    name:str = Field(description="Complete Name")
    email:str = Field(description="Email Address")
    age:int =  Field(description="Age")

structured_llm = llm.with_structured_output(ResponseStructure)
```

```python
res = structured_llm.invoke(f"Please give me only name , age and email from this text :{text}")
```

```python
class Movies(BaseModel):
    title:str = Field(description="Movie Title")
    Year:int = Field(description="Movie Release Year")


class AllMovies(BaseModel):
    movies:List[Movies]

movie_llm = llm.with_structured_output(Movies)
movies_llm = llm.with_structured_output(AllMovies)
```

```python
movie_llm.invoke("give me 1 sci fi recent movie")
```
